```python
r1 = chargement_donnee('../batchs', 'restaurant_1', 50, 60)
r2 = chargement_donnee('../batchs', 'restaurant_2', 50, 60)
```

```python
def clean(df3: pd.DataFrame) -> pd.DataFrame:
    df3.columns = df3.columns.str.lower().str.replace(' ','_')
    df3['order_date'] = pd.to_datetime(df3['order_date'])
    # CA
    df3['total_product_price'] = df3['quantity']*df3['product_price']
    df3['cash_in'] =  df3.groupby('order_number')['total_product_price'].transform(np.sum)
    df3 = df3.drop(columns=['item_name','quantity','product_price','total_products','total_product_price'])
    df3 = df3.drop_duplicates()
    df3 = df3.reset_index(drop=True)
    return df3
```

```python
r1 = clean(r1)

r1.head()
```

```python
# Nettoyage des données
def clean2(df2: pd.DataFrame) -> pd.DataFrame:
    df2 = df2.rename(columns={'order_id':'order_number'})
    df2.columns = df2.columns.str.lower().str.replace(' ','_')
    df2['order_date'] = pd.to_datetime(df2['order_date'])
    # CA
    df2['total_product_price'] = df2['quantity']*df2['product_price']
    df2['cash_in'] =  df2.groupby('order_id')['total_product_price'].transform(np.sum)
    df2 = df2.drop(columns=['item_name','quantity','product_price','total_products','total_product_price'])
    df2 = df2.drop_duplicates()
    df2 = df2.reset_index(drop=True)
    return df2
```

```python
r2 = clean2(r2)

r2.head()
```

```python
def merge(df2: pd.DataFrame, df3: pd.DataFrame)->pd.DataFrame:
    df4 = pd.concat([df2,df3], sort=True)
    df4 = df4.drop(columns='order_number')
    df4 = df4.sort_values('order_date')
    df4 = df4.reset_index(drop=True)
    return df4
```

```python
resto = merge(r1, r2)

resto.head()
```

```python
resto = resto.resample('1H', on='order_date').sum().reset_index()

resto.head()
```

```python
def etl(start_date : int, end_date : int) -> pd.DataFrame:
    r1 = chargement_donnee('../batchs', 'restaurant_1', start_date, end_date)
    r2 = chargement_donnee('../batchs', 'restaurant_2', start_date, end_date)
    r1 = clean(r1)
    r2 = clean2(r2)
    resto = merge(r1, r2)
    resto = resto.resample('1H', on='order_date').sum().reset_index()
    return resto
```

```python
#Jour de la semaine, jour, mois, année, lag
def prep(df: pd.DataFrame, date: str):
    
    df['jour'] = df[date].dt.weekday
    df = pd.get_dummies(df, columns=['jour'], drop_first= True)
    df['lag_1W'] = df['cash_in'].shift(24*7)
    df = df.dropna()
    df = df.reset_index()
    
    return df
#Regression linéaire / arbre de décision
```

```python
df = prep(df, 'order_date')

df.head(50)
```

```python
from sklearn.ensemble import RandomForestRegressor

simple_model = RandomForestRegressor(n_estimators= 10, random_state=42)
```

```python
simple_model.fit(df[['jour_1', 'jour_2', 'jour_3', 'jour_4', 'jour_5', 'jour_6', 'lag_1W']], df['cash_in'])
```
