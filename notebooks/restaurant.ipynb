```python
def chargement_donnee(dossier : str, fichier : str, start_date : int, end_date : int):
    #Récupérer le nom de tout les fichiers du dossier
    files = os.listdir(dossier)
    
    name = []
    for i in files:
        if i.find(fichier)!=-1:
            name.append(i)
    df = pd.DataFrame()
    #boucle pour chaque fichier fichier
    for i in name:
        if int(i[18:21]) >= start_date and int(i[18:21]) <= end_date:
            temp = pd.read_csv(path + '/' + i)
            df = pd.concat([df, temp])
    df['Order Date'] = pd.to_datetime(df['Order Date'])
    return df

df = chargement_donnee('../batchs', 'restaurant_', 50, 60)

df.head()
```

```python
df = df.resample('1H', on='Order Date').sum().reset_index()

df.head()
```
